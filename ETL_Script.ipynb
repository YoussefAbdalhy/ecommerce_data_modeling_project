# Data Transformation


### Date Dim

```python
start_date = '2010-01-01'
end_date = '2024-12-31'

date_range = pd.date_range(start=start_date, end=end_date)

date_dim = pd.DataFrame({'date': date_range})

date_dim['year'] = date_dim['date'].dt.year
date_dim['month'] = date_dim['date'].dt.month
date_dim['day'] = date_dim['date'].dt.day

# Modify day_of_week: Sunday=1, Monday=2, ...
date_dim['day_of_week'] = (date_dim['date'].dt.dayofweek + 2) % 7 + 1  

# Adjust the week number to start on Sunday
date_dim['week'] = date_dim['date'].dt.to_period('W-SUN').apply(lambda r: r.week)

date_dim['quarter'] = date_dim['date'].dt.quarter

# Modify is_weekend logic (Friday=6, Saturday=7)
date_dim['is_weekend'] = date_dim['day_of_week'].isin([6, 7])  

date_dim['month_name'] = date_dim['date'].dt.strftime('%B')
date_dim['day_name'] = date_dim['date'].dt.strftime('%A')

# Define seasons based on month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create the season column
date_dim['season'] = date_dim['month'].apply(get_season)


date_dim['date_id'] = date_dim['date'].dt.strftime('%Y%m%d').astype(int)
```

```python
date_dim.info()
```

### Time Dim

```python
hours = range(0, 24)  
minutes = range(0, 60) 
seconds = range(0, 60) 

# Create a list of all time combinations (hours, minutes, seconds)
time_combinations = [(h, m, s) for h in hours for m in minutes for s in seconds]

time_dim = pd.DataFrame(time_combinations, columns=['hour', 'minute', 'second'])

time_dim['time'] = time_dim.apply(lambda row: f"{row['hour']:02}:{row['minute']:02}:{row['second']:02}", axis=1)
time_dim['time'] = pd.to_datetime(time_dim['time'], format='%H:%M:%S')

time_dim['time_id'] = time_dim['time'].dt.strftime('%H%M%S').astype(int)
```

```python
time_dim.info()
```

### Orders Dim

```python
orders[orders.duplicated(subset=('order_id'))]
#no dupilcates 
```

```python
#chnage type of timestamps to be datetime
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['order_approved_date'] = pd.to_datetime(orders['order_approved_date']) 
orders['pickup_date'] = pd.to_datetime(orders['pickup_date'])
orders['delivered_date'] = pd.to_datetime(orders['delivered_date'])

#change type of estimated delievary date to be a date 
orders['estimated_time_delivery'] = pd.to_datetime(orders['estimated_time_delivery']).dt.date
```

```python
#Get Customer Internal Id instead of user name to make it easier in joining 
orders =orders.merge(customers, left_on = 'user_name',right_on = 'customer_name', how = 'left')

#only keep needed columns
orders = orders[['order_id','customer_internal_id','order_status','order_date','order_approved_date','pickup_date','delivered_date','estimated_time_delivery']]

#add order internal id (int) to make it easier while joning  
orders.insert(0,'order_internal_id',range(1,len(orders)+1))
```

```python
#get the date out of timestamp coulmns
orders['order_date_date'] = orders['order_date'].dt.date
orders['order_approved_date_date'] = orders['order_approved_date'].dt.date
orders['pickup_date_date'] = orders['pickup_date'].dt.date
orders['delivered_date_date'] = orders['delivered_date'].dt.date
orders['estimated_time_delivery_date'] = orders['estimated_time_delivery']

orders['order_date_date'] = pd.to_datetime(orders['order_date_date'])
orders['order_approved_date_date'] = pd.to_datetime(orders['order_approved_date_date'])
orders['pickup_date_date'] = pd.to_datetime(orders['pickup_date_date'])
orders['delivered_date_date'] = pd.to_datetime(orders['delivered_date_date'])
orders['estimated_time_delivery_date'] = pd.to_datetime(orders['estimated_time_delivery_date'])


#get time out from timestamp columns 
orders['order_date_time'] = orders['order_date'].dt.strftime('%H:%M:%S')
orders['order_approved_date_time'] = orders['order_approved_date'].dt.strftime('%H:%M:%S')
orders['pickup_date_time'] = orders['pickup_date'].dt.strftime('%H:%M:%S')
orders['delivered_date_time'] = orders['delivered_date'].dt.strftime('%H:%M:%S')
```

```python
#merge with date dim to get date ids 
orders = orders.merge(date_dim, left_on='order_date_date', right_on='date', how='left') \
               .merge(date_dim, left_on='order_approved_date_date', right_on='date', how='left', suffixes=('_order', '_approved')) \
               .merge(date_dim, left_on='pickup_date_date', right_on='date', how='left', suffixes=('_approved', '_pickup')) \
               .merge(date_dim, left_on='delivered_date_date', right_on='date', how='left', suffixes=('_pickup', '_delivered')) \
               .merge(date_dim, left_on='estimated_time_delivery_date', right_on='date', how='left',suffixes=('estimated', 'estimated'))
time_dim['time'] = time_dim['time'].dt.strftime('%H:%M:%S')

#mere with time dim to get time ids 
orders = orders.merge(time_dim, left_on='order_date_time', right_on='time', how='left') \
               .merge(time_dim, left_on='order_approved_date_time', right_on='time', how='left', suffixes=('_order_time', '_approved_time')) \
               .merge(time_dim, left_on='pickup_date_time', right_on='time', how='left', suffixes=('_approved_time', '_pickup_time')) \
               .merge(time_dim, left_on='delivered_date_time', right_on='time', how='left', suffixes=('_pickup_time', '_delivered_time'))

```

```python
#keep needed columns 
orders = orders[['order_internal_id','order_id','customer_internal_id','order_status', 'date_id_order','time_id_order_time', 'date_id_approved','time_id_approved_time',  'date_id_pickup','time_id_pickup_time', 'date_id_delivered', 'time_id_delivered_time','date_id']]
```

```python
#rename to make it easier to read
orders = orders.rename(columns = {'date_id_order':'order_date_id',
                                  'time_id_order_time':'order_date_time_id',
                                 'date_id_approved':'order_approved_at_date_id',
                                 'time_id_approved_time':'order_approved_at_time_id',
                                 'date_id_pickup':'pickup_date_id',
                                 'time_id_pickup_time':'pickup_time_id',
                                 'date_id_delivered':'delivered_at_date_id',
                                 'time_id_delivered_time':'delivered_at_time_id',
                                 'date_id':'estimated_delivery_date_id'})
```

### Order Items Dim

```python
order_items[order_items.duplicated(subset = ['order_id','product_id','order_item_id'])]
#no duplicates
```

```python
#get order internal id 
order_items = order_items.merge(orders, on ='order_id',how='left')

#get product internal id
order_items = order_items.merge(products, on ='product_id',how='left')

#get seller internal id
order_items = order_items.merge(sellers, on ='seller_id',how='left')

#keep needed columns 
order_items = order_items[['order_internal_id','order_item_id','product_internal_id','seller_internal_id','pickup_limit_date','price','shipping_cost']]

# change pickup limit date to date time
order_items['pickup_limit_date'] = pd.to_datetime(order_items['pickup_limit_date'])

#get the date from timestamp column 
order_items['pickup_limit_date_date'] = order_items['pickup_limit_date'].dt.date
order_items['pickup_limit_date_date']  = pd.to_datetime(order_items['pickup_limit_date_date'])

#get the time from timestamp column 
order_items['pickup_limit_date_time'] = order_items['pickup_limit_date'].dt.strftime('%H:%M:%S')

#merge to get date id 
order_items = order_items.merge(date_dim, left_on= 'pickup_limit_date_date', right_on ='date')

#merge to get time id 
order_items = order_items.merge(time_dim, left_on= 'pickup_limit_date_time', right_on ='time')

#keep needed columns 
order_items = order_items[['order_internal_id','order_item_id','product_internal_id','seller_internal_id','date_id','time_id','price','shipping_cost']]

#rename columns
order_items = order_items.rename(columns = {'date_id':'pickup_limit_date_id','time_id':'pickup_limit_time_id'})
```

### Feedbacks Dim

```python
feedbacks[feedbacks['order_id'].duplicated()] 
# there are duplicate feedback scores for the same order 
```

```python
#change the type of timestamp columns to be date time
feedbacks['feedback_form_sent_date'] = pd.to_datetime(feedbacks['feedback_form_sent_date']).dt.date
feedbacks['feedback_answer_date'] = pd.to_datetime(feedbacks['feedback_answer_date'])
```

```python
#get date out of timestamp columns 
feedbacks['feedback_answer_date_date'] = feedbacks['feedback_answer_date'].dt.date

feedbacks['feedback_form_sent_date']  = pd.to_datetime(feedbacks['feedback_form_sent_date'])
feedbacks['feedback_answer_date_date']  = pd.to_datetime(feedbacks['feedback_answer_date_date'])


#get time out of timestamp columns 
feedbacks['feedback_answer_date_time'] = feedbacks['feedback_answer_date'].dt.strftime('%H:%M:%S')

#merge to get date_ids 
feedbacks = feedbacks.merge(date_dim, left_on = 'feedback_form_sent_date', right_on= 'date',how = 'left' ) \
                    .merge (date_dim, left_on = 'feedback_answer_date_date', right_on= 'date',how = 'left',suffixes = ('_sent','_answer'))

#merge to get time_id
feedbacks = feedbacks.merge(time_dim, left_on = 'feedback_answer_date_time', right_on= 'time',how = 'left' )

# join with orders to add order internal id 
feedbacks = feedbacks.merge(orders, on = 'order_id', how ='left')

#keep needed columns 
feedbacks = feedbacks[['feedback_id','order_internal_id','feedback_score','date_id_sent','date_id_answer','time_id']]

#rename columns to make it easier 
feedbacks =feedbacks.rename(columns = {'date_id_sent':'feedback_form_sent_date_id',
                                      'date_id_answer':'feedback_form_answer_date_id',
                                      'time_id':'feedback_form_answer_time_id'})

# Add feedback internal id (int) to make it easier while joning   
feedbacks.insert (0,'feedback_internal_id',range(1,len(feedbacks)+1))
```

### Sales Fact

```python
sales_fact = order_items.groupby(['order_internal_id','seller_internal_id','product_internal_id']).agg(
                                                qyt = ('product_internal_id','count'),
                                                total_price = ('price','sum'),
                                                total_shipping_cost = ('shipping_cost','sum')).reset_index()
```

```python
sales_fact.info()
```

### Payments Fact 

```python
payment_summary_fact = payments.groupby(['order_internal_id','payment_type']).agg(
                        payments_count = ('payment_type','count'),
                        total_payment_installments =('payment_installments','sum'),
                        total_payment_value = ('payment_value','sum')).reset_index()

payment_summary_fact['total_payment_installments'] = payment_summary_fact['total_payment_installments'].where(payment_summary_fact['payment_type'] == 'credit_card', pd.NA).astype('Int64')
```

```python
try:
    with engine.connect() as connection:
        dataframes = [date_dim, time_dim, sellers, customers, products, orders, feedbacks, order_items, payments, sales_fact, payment_summary_fact]  
        table_names = ['date_dim', 'time_dim', 'sellers_dim', 'customers_dim', 'products_dim', 'orders_dim', 'feedbacks', 'order_items_dim', 'payments_dim', 'sales_fact','payment_summary_fact']  

        # Truncate and insert data
        for df, table_name in zip(dataframes, table_names):
            connection.execute(f"truncate table {table_name} CASCADE")
            df.to_sql(table_name, engine, if_exists='append', index=False)
        print('Data Loaded Successfully')
        
except OperationalError as e:
    print(f"Loading Data Failed: {e}")
```
